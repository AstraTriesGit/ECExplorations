## Periodic Signals
#### Square Wave
of amplitude A, fundamental frequency w0 (rad/s), duty cycle rho (_the fraction of each period where signal is +ve_)

```julia
using Plots
using ECExplorations.Signals

A = 1
w0 = 0.2
rho = 0.5

square_wave = [A* square(ti, w0, rho) for ti in t]

plot(t, square_wave, 
     label="Custom Square Wave", 
     title="Custom Square Wave Plot", 
     xlabel="Time", 
     ylabel="Amplitude",
     linewidth=2)
```

#### Triangle Wave
of amplitude A, fundamental frequency w0 (here just in s-1, sorry Haykin), width W, period T, first max value occurs at t = WT.

```julia
using Plots
using ECExplorations.Signals.Triangle

A = 2
w0 = 0.4
t = 0:0.001:2

wave1 = [A * Triangle.asin_t(ti) for ti in t]
plot(t, wave1, label="asin method", title="Triangle Wave", xlabel="Time", ylabel="Amplitude")
```
